```python
from pathlib import Path
import pandas as pd
```

```python
repo_root = Path.cwd()
if repo_root.name == "Combined_datasets":
    repo_root = repo_root.parent
```

## Biometric

```python
bio_dir = repo_root / "api_data_aadhar_biometric"
bdf1 = pd.read_csv(bio_dir / "api_data_aadhar_biometric_0_500000.csv")
bdf2 = pd.read_csv(bio_dir / "api_data_aadhar_biometric_500000_1000000.csv")
bdf3 = pd.read_csv(bio_dir / "api_data_aadhar_biometric_1000000_1500000.csv")
bdf4 = pd.read_csv(bio_dir / "api_data_aadhar_biometric_1500000_1861108.csv")

output_path = repo_root / "Combined_datasets" / "api_data_aadhar_biometric_combined.csv"
output_path.parent.mkdir(parents=True, exist_ok=True)

bdf = pd.concat([bdf1, bdf2, bdf3, bdf4], ignore_index=True)
bdf.to_csv(output_path, index=False)
output_path
```

## Demographic

```python
demo_dir = repo_root / "api_data_aadhar_demographic"
ddf1 = pd.read_csv(demo_dir / "api_data_aadhar_demographic_0_500000.csv")
ddf2 = pd.read_csv(demo_dir / "api_data_aadhar_demographic_500000_1000000.csv")
ddf3 = pd.read_csv(demo_dir / "api_data_aadhar_demographic_1000000_1500000.csv")
ddf4 = pd.read_csv(demo_dir / "api_data_aadhar_demographic_1500000_2000000.csv")
ddf5 = pd.read_csv(demo_dir / "api_data_aadhar_demographic_2000000_2071700.csv")

demo_output = repo_root / "Combined_datasets" / "api_data_aadhar_demographic_combined.csv"
demo_output.parent.mkdir(parents=True, exist_ok=True)

ddf = pd.concat([ddf1, ddf2, ddf3, ddf4, ddf5], ignore_index=True)
ddf.to_csv(demo_output, index=False)
demo_output
```

## Enrolment

```python
enrol_dir = repo_root / "api_data_aadhar_enrolment"
edf1 = pd.read_csv(enrol_dir / "api_data_aadhar_enrolment_0_500000.csv")
edf2 = pd.read_csv(enrol_dir / "api_data_aadhar_enrolment_500000_1000000.csv")
edf3 = pd.read_csv(enrol_dir / "api_data_aadhar_enrolment_1000000_1006029.csv")

enrol_output = repo_root / "Combined_datasets" / "api_data_aadhar_enrolment_combined.csv"
enrol_output.parent.mkdir(parents=True, exist_ok=True)

edf = pd.concat([edf1, edf2, edf3], ignore_index=True)
edf.to_csv(enrol_output, index=False)
enrol_output
```
